```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("SuperMarket Analysis.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.shape
```

```python
city_sales = df.groupby("City").sum()
city_sales
```

```python
city_sales
```

```python
city_sales["Sales"]
```

```python
import matplotlib.pyplot as plt

city_sales["Sales"].plot(kind="bar", title="City wise Sales")
plt.xlabel("City")
plt.ylabel("Sales")
plt.show()
```
